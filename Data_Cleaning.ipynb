```python
import pandas as pd
import numpy as np

# Import the Dataset file
df = pd.read_csv("Luxury_Housing_Bangalore.csv").copy()
# Print the first 5 rows of the dataframe
print(df.head())


```

```python
df.dtypes
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].astype('Int64')
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].astype('object').str.replace('₹','').str.replace('Cr','').astype('float')
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].round(2).astype('float')
df['Connectivity_Score'] = df['Connectivity_Score'].round(1).astype('float')
df['Amenity_Score'] = df['Amenity_Score'].round(1).astype('float')
df['Locality_Infra_Score'] = df['Locality_Infra_Score'].round(1).astype('float')
df['Configuration'] = df['Configuration'].str.replace(r'(?i)bhk\+?', '', regex=True).str.strip()
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
df.isna().sum()
```

```python
df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median(), inplace=True)
df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median(), inplace=True)
df['Amenity_Score'].fillna(df['Amenity_Score'].median(), inplace=True)
df['Buyer_Comments'].fillna('No Comments', inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
import psycopg2
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg2://postgres:Test123@localhost:5432/Luxury_Housing_Bangalore')
engine.connect()
df.to_sql('Luxury_Housing_Bangalore', engine, if_exists='replace', index=False)
load_dotenv()
df.dtypes

```

```python
def get_db_connection_params():
    return {
        'dbname': os.getenv('DB_NAME'),
        'user': os.getenv('DB_USER'),
        'password': os.getenv('DB_PASSWORD'),
        'host': os.getenv('DB_HOST'),
        'port': os.getenv('DB_PORT')
    }
print(get_db_connection_params())

def test_psycopg2_connection():
    params = get_db_connection_params()
    try:
        conn = psycopg2.connect(**params)
        print("Connection successful")
        conn.close()
    except Exception as e:
        print(f"Connection failed: {e}")
test_psycopg2_connection()

```

```python
def create_table():
    db = psycopg2.connect(**get_db_connection_params())
    cursor = db.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS Luxury_Housing_Bangalore (
            Project_Name VARCHAR,
            Location VARCHAR,
            Configuration VARCHAR,
            Unit_Size_Sqft INT,
            Ticket_Price_Cr FLOAT,
            Connectivity_Score FLOAT,
            Amenity_Score FLOAT,
            Locality_Infra_Score FLOAT,
            Buyer_Comments TEXT
        )
    """)
    db.commit()
    cursor.close()
    db.close()
create_table()
```
